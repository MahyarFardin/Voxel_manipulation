```python
PROJECT_ROOT = Path(".").resolve().parent
#PROJECT_ROOT = Path("/home/salman/rai_workspace/Voxel_manipulation_v2/pose_estimation")

#MESH_INPUT  = PROJECT_ROOT / "environments/mesh_dataset/lamp/lamp_0001.ply"
#MESH_OUTPUT = PROJECT_ROOT / "megapose_objects/lamp/mesh.ply"

#MESH_INPUT  = PROJECT_ROOT / "environments/mesh_dataset/cone/cone_0001.ply"
#MESH_OUTPUT = PROJECT_ROOT / "megapose_objects/cone/mesh.ply"

#MESH_INPUT  = PROJECT_ROOT / "environments/mesh_dataset/redCup/RedCup_25k_tex.ply"
#MESH_OUTPUT = PROJECT_ROOT / "megapose_objects/redCup/mesh.ply"

#MESH_INPUT  = PROJECT_ROOT / "environments/mesh_dataset/bottle/Sprudelflasche_25k_tex.ply"
#MESH_OUTPUT = PROJECT_ROOT / "megapose_objects/bottle/mesh.ply"

#MESH_INPUT  = PROJECT_ROOT / "environments/mesh_dataset/crackerBox/visual_hull_refined_smoothed.ply"
#MESH_OUTPUT = PROJECT_ROOT / "megapose_objects/crackerBox/mesh.ply"

MESH_INPUT  = PROJECT_ROOT / "environments/mesh_dataset/deodorant/Deodorant_25k_tex.ply"
MESH_OUTPUT = PROJECT_ROOT / "megapose_objects/deodorant/mesh.ply"

#MESH_SCALE = 0.000271579    # meshscale lamp from the .g scene file
#MESH_SCALE = 0.00708861     # meshscale cone from the .g scene file
#MESH_SCALE = 0.000943656    # meshscale redCup from the .g scene file
#MESH_SCALE = 0.000871165    # meshscale bottle from the .g scene file
#MESH_SCALE = 0.542528       # meshscale crackerBox from the .g scene file
MESH_SCALE = 0.00147527     # meshscale deodorant from the .g scene file


MESH_OUTPUT.parent.mkdir(parents=True, exist_ok=True)
```

```python
mesh = trimesh.load(str(MESH_INPUT))
print(f"Loaded:  {len(mesh.vertices)} vertices, {len(mesh.faces)} faces")
print(f"Bounds before scaling:\n  {mesh.bounds}")
print(f"Extents before scaling: {mesh.extents}")
```

```python
mesh.apply_scale(MESH_SCALE)

print(f"Bounds after scaling (meters):\n  {mesh.bounds}")
print(f"Extents after scaling (meters): {mesh.extents}")
print(f"Longest dimension: {mesh.extents.max():.4f} m")
```

```python
mesh.export(str(MESH_OUTPUT))
print(f"Saved scaled mesh -> {MESH_OUTPUT}")
```
