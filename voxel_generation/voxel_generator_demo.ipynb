```python
from VoxelGenerator import VoxelGenerator


def main():
    # Create the generator object
    generator = VoxelGenerator(
        base_file="base.g",      # Path to the base .g file that contains the initial object/setup
        output_root="./data",    # Root directory where generated voxel folders will be created
        step=0.1,                # Distance between neighboring voxel centers
        default_size=0.1         # Default side length of each cube
    )

    # Generate voxel structure files
    created_files = generator.generate_voxel_files(
        folder_name="voxels",    # Subfolder inside output_root -> ./data/voxels/
        num_voxels=100,          # Number of .g files to generate
        n_range=(1, 10),         # Each structure will contain a random number of cubes in [1, 10)
        size=0.1                 # Cube size; if None, default_size is used
    )

    # Visualize all generated voxel files from that folder
    generator.visualize_folder(
        folder_name="voxels",    # Folder to visualize
        spacing_low=1,           # Minimum random offset when placing each structure in the viewer
        spacing_high=10          # Maximum random offset
    )


if __name__ == "__main__":
    main()
```
