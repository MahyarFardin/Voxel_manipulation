```python
import robotic as ry
import numpy as np
import random
from glob import glob
import os
```

```python
def add_random_joints(color, size, file_path="base.g", n=np.random.randint(1, 10)):
    pixles = [[0, 0, 0]]
    base = f"""base {{}}

cube0 (base) {{ 
    shape: box, 
    size: [{size}, {size}, {size}], 
    color: {color}, 
    X: "t(0 0 0)" 
}}"""

    for i in range(1, n):
        parent = random.choice(pixles)
        new_popsition = parent.copy()

        while new_popsition in pixles:
            new_popsition[np.random.randint(0, 3)] += .1

        base += f"""
joint{i}(cube{i-1}):{{
  joint: rigid,
  pre: "T t({parent[0]} {parent[1]} {parent[2]})" 
}}

cube{i}(joint{i}):{{ 
  shape: box, 
  size: [{size}, {size}, {size}], 
  color: {color},
  position: [{new_popsition[0]}, {new_popsition[1]}, {new_popsition[2]}],
  X: "t({new_popsition[0]} {new_popsition[1]} {new_popsition[2]})" 
}}
"""
        pixles.append(new_popsition)
    with open("./data/"+str(int(np.random.rand()*1000))+".g", 'w', encoding='utf-8') as f: f.write(base)
```

```python
# define this loop as much as you want 
os.makedirs("./data", exist_ok=True)

for i in range(100):
    add_random_joints(color=np.random.randint(0, 2, (3, )), size=.1)
```

```python
# run this cell to visualize all the voxels generated
p = [0, 0, 0]
C = ry.Config()

for f in glob("./data/*.g"):
    r = np.random.randint(1, 10, (3,))
    C.addFile(f).setPosition(p+r)
```
